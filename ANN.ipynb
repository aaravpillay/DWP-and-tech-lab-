```python
def gd(x,y):
  m_curr=b_curr=0 #initialize m and b
  n=len(x)
  lr=0.01

  for i in range(1000): #no. of baby steps
    y_pred=m_curr*x+b_curr
    loss=(1/n)*sum([val**2 for val in (y-y_pred)]) # corrected the assignment to loss, changed - to =.
    md=-(2/n)*sum(x*(y-y_pred))
    bd=-(2/n)*sum(y-y_pred)
    m_curr=m_curr-lr*md # corrected the assignment to m_curr, not m_crr.
    b_curr=b_curr-lr*bd
    print("m {},b {},cost {} iteration {}".format(m_curr,b_curr,loss,i))

import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gd(x,y)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pdfrom
from sklearn.datasets import load_iris
#Load dataset
iris=load_iris()
x=iris.data
y=iris.target
#Convert to binary classification
y=(y==0).astype(int)
#Split into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#Standardize the features
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
#Initialize perceptron model
perceptron=Perceptron(max_iter=1000,eta0=0.01,random_state=42)
#Train the model
perceptron.fit(x_train_scaled,y_train)
#Make predictions on the test set
y_pred=perceptron.predict(x_test_scaled)
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)

#Split into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(y_test)
df=pd.DataFrame(y_test)
print(df.shape)
#Standardize the features
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
#Initialize perceptron model
perceptron=Perceptron(max_iter=1000,eta0=0.01,random_state=42)
#Train the model
perceptron.fit(x_train_scaled,y_train)
#Make predictions on the test set
y_pred=perceptron.predict(x_test_scaled)

print("Accuracy:",accuracy_score(y_test,y_pred))
```
